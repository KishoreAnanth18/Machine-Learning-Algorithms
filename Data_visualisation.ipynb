```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line plot

x=[1,2,3,4,5]
y0=[2,4,6,8,7]
y1=[4,5,8,7,2]
plt.xlabel("X Axis", fontsize = 12)
plt.ylabel("Y Axis", fontsize = 12)
plt.title("Line plot", fontsize = 16)
plt.plot(x,y0,color="red",label="Line Graph0")
plt.plot(x,y1,color="green",label="Line Graph1")
plt.legend(loc=2)
# plt.grid()
# plt.axis('off') #to remove axis
plt.show()
```

```python
# Bar graph

x=['A','B','C','D','E']
y=[1,2,3,4,5]
plt.bar(x, y, color ='blue', width = 0.4)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bar graph")
plt.show()
```

```python
# Scatter plot

x=[1,2,3,4,5]
y0=[2,4,6,8,7]
y1=[4,5,8,7,2]
plt.xlabel("X Axis", fontsize = 12)
plt.ylabel("Y Axis", fontsize = 12)
plt.title("Scatter plot", fontsize = 16)
plt.scatter(x,y0,color="red",label="First graph")
plt.scatter(x,y1,color="green",label="Second graph")
plt.legend(loc=2)
plt.plot(x,y1,color="blue")
plt.plot(x,y0,color="black")
# plt.grid()
# plt.axis('off') #to remove axis
plt.show()
```

```python
# Histogram

sample=np.random.randint(10,100,20)
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(sample, bins = [0, 25, 50, 75, 100],width=1)
 
# Show plot
plt.show()
```

```python
# Pie chart

plt.figure(figsize=(7,7))
slices = [10,20,50,30,45]
act = ["A","B","C","D","E"]
cols = ['red','blue','green','yellow','pink']
plt.pie(slices,labels=act,colors=cols,autopct="%1.2f%%",explode=(0,0.5,0,0,0))
plt.show()
```
