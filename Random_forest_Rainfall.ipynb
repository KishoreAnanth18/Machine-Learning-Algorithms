```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
df = pd.read_csv('/content/district wise rainfall normal.csv')
```

```python
df.head()
```

```python
df.drop(['STATE_UT_NAME','DISTRICT'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df = df+0.00001*np.random.rand(641, 17)
```

```python
sns.pairplot(df,hue='ANNUAL',palette='Oranges')
```

```python
X = df.drop('ANNUAL',axis=1) 
Y = df['ANNUAL']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
```

```python
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(X_train, Y_train)
```

```python
rfr_pred = rfr.predict(X_test)
```

```python
rfr_pred
```

```python
print("Accuracy:\n")
r2_score(Y_test,rfr_pred)
```
